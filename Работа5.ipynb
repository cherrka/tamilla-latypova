6. Разработайте модуль для классификации текстов по категориям с использованием zero-shot classification.

```python
from transformers import pipeline

# Инициализируем pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

def classify_text(text, candidate_labels, multi_label=False):
    """
    Классификация текста с использованием zero-shot learning.

    :param text: Строка — входной текст
    :param candidate_labels: Список — категории для классификации
    :param multi_label: True — если текст может относиться к нескольким категориям
    :return: Словарь с метками и вероятностями
    """
    result = classifier(text, candidate_labels, multi_label=multi_label)
    return dict(zip(result['labels'], result['scores']))

# Пример
text = "The new Tesla model is an impressive electric vehicle with autopilot features."
labels = ["technology", "sports", "politics", "automobile", "entertainment"]

result = classify_text(text, labels)
print("Классификация:")
for label, score in result.items():
    print(f"{label}: {score:.2f}")
```

8. Напишите скрипт, который генерирует аннотацию на основе произвольного текста, применяя модель t5-small

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Загружаем модель и токенизатор
model_name = "t5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

def generate_summary(text, max_length=80, min_length=20):
    # Добавляем префикс задачи
    input_text = "summarize: " + text.strip()

    # Токенизация
    input_ids = tokenizer.encode(input_text, return_tensors="pt", truncation=True)

    # Генерация аннотации
    summary_ids = model.generate(input_ids,
                                  max_length=max_length,
                                  min_length=min_length,
                                  length_penalty=2.0,
                                  num_beams=4,
                                  early_stopping=True)

    # Декодирование результата
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Пример использования
text = """
Machine learning is a method of data analysis that automates analytical model building.
It is a branch of artificial intelligence based on the idea that systems can learn from data,
identify patterns and make decisions with minimal human intervention.
"""

summary = generate_summary(text)
print("Аннотация:\n", summary)
```

9. Реализуйте модуль для извлечения векторных представлений слов из текста с помощью модели bert-base-uncased.

```python
import torch
from transformers import BertTokenizer, BertModel

# Загружаем модель и токенизатор
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
model.eval()

def get_token_embeddings(text):
    # Токенизация текста
    inputs = tokenizer(text, return_tensors="pt", return_attention_mask=True)
    with torch.no_grad():
        outputs = model(**inputs)

    # Последний скрытый слой
    last_hidden_state = outputs.last_hidden_state  # [batch_size, seq_len, hidden_size]

    tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])
    embeddings = last_hidden_state[0]  # [seq_len, hidden_size]

    return list(zip(tokens, embeddings))

# Пример
text = "BERT is a powerful language model."
token_vectors = get_token_embeddings(text)

# Печать: токен + первые 5 значений вектора
for token, vec in token_vectors:
    print(f"{token:12s} -> {vec[:5].tolist()}")
```

10. Разработайте систему для генерации продолжения текста на основе заданного начала, используя модель gpt2.

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Загружаем модель и токенизатор
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()

def generate_text(prompt, max_length=100, temperature=1.0, top_k=50, top_p=0.95):
    # Токенизация
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Генерация
    with torch.no_grad():
        output_ids = model.generate(
            input_ids,
            max_length=max_length,
            temperature=temperature,
            top_k=top_k,
            top_p=top_p,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )

    # Декодирование
    generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return generated_text

# Пример
prompt = "Once upon a time in a distant kingdom"
generated = generate_text(prompt)
print("Сгенерированный текст:\n")
print(generated)
```

11. Создайте приложение для заполнения пропусков в предложениях на русском и английском языках с помощью модели bert-base-multilingual-cased.

```python
from transformers import BertTokenizer, BertForMaskedLM
import torch

# Загружаем токенизатор и модель
model_name = "bert-base-multilingual-cased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForMaskedLM.from_pretrained(model_name)
model.eval()

def fill_mask(text_with_mask, top_k=5):
    """
    Заполняет пропущенное слово, обозначенное как [MASK]
    :param text_with_mask: Текст с одним [MASK]
    :param top_k: Количество предсказаний
    """
    # Токенизация
    input_ids = tokenizer.encode(text_with_mask, return_tensors="pt")
    mask_token_index = torch.where(input_ids == tokenizer.mask_token_id)[1]

    # Прогноз
    with torch.no_grad():
        output = model(input_ids)
    logits = output.logits

    mask_token_logits = logits[0, mask_token_index, :]
    top_k_ids = torch.topk(mask_token_logits, top_k, dim=1).indices[0].tolist()

    print(f"\nВарианты заполнения для: {text_with_mask}")
    for token_id in top_k_ids:
        token = tokenizer.decode([token_id])
        filled_text = text_with_mask.replace(tokenizer.mask_token, token)
        print(f"- {filled_text}")

# Примеры:
fill_mask("I love [MASK] food.")
fill_mask("Она пошла в [MASK] вечером.")
```

12. Напишите скрипт, который определяет сходство между двумя предложениями, применяя модель sentence-transformers/paraphrase-MiniLM-L6-v2.

```python
from sentence_transformers import SentenceTransformer, util

# Загружаем модель
model = SentenceTransformer('sentence-transformers/paraphrase-MiniLM-L6-v2')

def sentence_similarity(sent1, sent2):
    # Получение эмбеддингов
    embeddings = model.encode([sent1, sent2], convert_to_tensor=True)

    # Вычисление косинусного сходства
    similarity = util.pytorch_cos_sim(embeddings[0], embeddings[1]).item()
    return similarity

# Пример
s1 = "A man is playing a guitar."
s2 = "Someone is playing music."

score = sentence_similarity(s1, s2)
print(f"Сходство между предложениями: {score:.4f}")
```

14. Создайте систему для идентификации языка текста, применяя модель distilbert-base-multilingual-cased-language-detection.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

# Загружаем модель и токенизатор
model_name = "papluca/xlm-roberta-base-language-detection"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Список языков
labels = [
    "af", "am", "ar", "as", "az", "be", "bem", "bg", "bn", "br", "bs", "ca", "cs", "cy", "da", "de", "dz",
    "el", "en", "eo", "es", "et", "eu", "fa", "fi", "fo", "fr", "ga", "gl", "gu", "he", "hi", "hr", "ht",
    "hu", "hy", "id", "is", "it", "ja", "jv", "ka", "kk", "km", "kn", "ko", "ku", "ky", "la", "lo", "lt",
    "lv", "mg", "mk", "ml", "mn", "mr", "ms", "mt", "ne", "nl", "no", "pa", "pl", "ps", "pt", "qu", "ro",
    "ru", "rw", "sd", "sh", "si", "sk", "sl", "so", "sq", "sr", "sv", "sw", "ta", "te", "th", "tl", "tr",
    "uk", "ur", "uz", "vi", "wa", "xh", "zh", "zu"
]

def detect_language(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
        probs = F.softmax(outputs.logits, dim=1)
        top_idx = torch.argmax(probs, dim=1).item()
        language = labels[top_idx]
        confidence = probs[0][top_idx].item()
        return language, round(confidence, 4)

# Пример
texts = [
    "Bonjour, comment allez-vous?",
    "Привет, как дела?",
    "This is a language detection test."
]

for text in texts:
    lang, conf = detect_language(text)
    print(f"Текст: {text}\n")
```

17. Реализуйте систему для выявления фактов и утверждений в тексте с помощью модели roberta-large-openai-detector.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

# Загрузка модели
model_name = "roberta-large-openai-detector"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

def detect_generated_text(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
        probs = F.softmax(outputs.logits, dim=1)
        real_prob = probs[0][0].item()
        generated_prob = probs[0][1].item()
        result = "Generated" if generated_prob > real_prob else "Human-written"
        return result, round(real_prob, 4), round(generated_prob, 4)

# Пример
text = "The Eiffel Tower was built in 1889 and stands in Paris."
result, real, fake = detect_generated_text(text)
print(f"Результат: {result} (real: {real}, generated: {fake})")
```

18. Создайте приложение для перефразирования текста, используя модель tuner007/pegasus_paraphrase.

```python
from transformers import PegasusTokenizer, PegasusForConditionalGeneration
import torch

# Загружаем токенизатор и модель
model_name = "tuner007/pegasus_paraphrase"
tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name)

def paraphrase(text, num_return_sequences=3, num_beams=10):
    """
    Перефразирование текста с использованием модели Pegasus.
    :param text: Строка — исходный текст
    :param num_return_sequences: Количество вариантов перефразировки
    :param num_beams: Количество beam'ов для генерации
    :return: Список перефразированных вариантов
    """
    inputs = tokenizer([text], truncation=True, padding="longest", return_tensors="pt")
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_length=60,
            num_beams=num_beams,
            num_return_sequences=num_return_sequences,
            temperature=1.5
        )
    return tokenizer.batch_decode(outputs, skip_special_tokens=True)

# Пример использования
original_text = "Machine learning is a method of data analysis that automates analytical model building."

paraphrased_versions = paraphrase(original_text)
print("Исходный текст:\n", original_text)
print("\nВарианты перефразирования:")
for i, version in enumerate(paraphrased_versions, 1):
    print(f"{i}) {version}")
```

20. Напишите скрипт, который определяет политическую окраску текста, применяя модель bhadresh-savani/distilbert-base-uncased-emotion.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Загружаем модель и токенизатор
model_name = "bhadresh-savani/distilbert-base-uncased-emotion"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Создаем пайплайн для классификации эмоций
emotion_classifier = pipeline("text-classification", model=model, tokenizer=tokenizer, return_all_scores=True)

def detect_emotion(text):
    emotions = emotion_classifier(text)[0]
    sorted_emotions = sorted(emotions, key=lambda x: x["score"], reverse=True)
    return sorted_emotions

# Пример текста
text = "The government must take action to protect workers' rights and raise the minimum wage."

emotions = detect_emotion(text)
print("Эмоциональная окраска текста:")
for e in emotions:
    print(f"{e['label']}: {round(e['score'], 4)}")
```

25. Реализуйте систему для перевода текстов с английского на испанский, используя модель Helsinki-NLP/opus-mt-en-es.

```python
from transformers import MarianMTModel, MarianTokenizer

# Загружаем модель и токенизатор для перевода с английского на испанский
model_name = "Helsinki-NLP/opus-mt-en-es"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_text(text, src_lang="en", tgt_lang="es"):
    """
    Перевод текста с одного языка на другой.
    :param text: Текст для перевода
    :param src_lang: Исходный язык
    :param tgt_lang: Язык перевода
    :return: Переведённый текст
    """
    # Токенизация входного текста
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

    # Генерация перевода
    translated = model.generate(**inputs)

    # Декодирование перевода
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translated_text

# Пример текста на английском языке
text = "Hello, how are you today? I hope you're doing well!"

# Перевод текста с английского на испанский
translated_text = translate_text(text)
print("Перевод на испанский:", translated_text)
```

27. Напишите скрипт, который извлекает сущности и отношения из текста с помощью модели nlpaueb/legal-bert-base-uncased.

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

# Загружаем модель и токенизатор для извлечения сущностей
model_name = "nlpaueb/legal-bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Создаём пайплайн для извлечения сущностей
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")

def extract_entities_and_relations(text):
    """
    Извлекает сущности и отношения из текста.
    :param text: Текст для извлечения сущностей
    :return: Сущности и их отношения в тексте
    """
    entities = ner_pipeline(text)

    # Выводим сущности и их типы
    for entity in entities:
        print(f"Сущность: {entity['word']}, Тип: {entity['entity_group']}, Уверенность: {entity['score']:.3f}")

    return entities

# Пример юридического текста
text = """
On March 10, 2020, John Doe entered into an agreement with XYZ Corp to provide services.
XYZ Corp will pay John Doe $100,000 for the completion of the work by July 15, 2020.
"""

# Извлечение сущностей и отношений из текста
entities = extract_entities_and_relations(text)
```

32. Разработайте приложение для перефразирования текстов с сохранением смысла, применяя модель tuner007/pegasus_paraphrase.

```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer
import torch

# Загрузка модели и токенизатора
model_name = "tuner007/pegasus_paraphrase"
tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name)

def paraphrase_text(text):
    """
    Функция для перефразирования текста с сохранением смысла.
    :param text: Исходный текст
    :return: Перефразированный текст
    """
    # Токенизация текста
    inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)

    # Генерация перефразированного текста
    with torch.no_grad():
        summary_ids = model.generate(inputs['input_ids'], num_beams=5, num_return_sequences=3, length_penalty=1.0)

    # Декодирование результата в текст
    paraphrases = [tokenizer.decode(g, skip_special_tokens=True) for g in summary_ids]

    return paraphrases

# Пример текста для перефразирования
text = "The quick brown fox jumps over the lazy dog."

# Перефразирование текста
paraphrased_texts = paraphrase_text(text)
print("Перефразированные варианты текста:")
for i, paraphrase in enumerate(paraphrased_texts, 1):
    print(f"{i}: {paraphrase}")
```

33. Реализуйте систему для определения степени достоверности информации в текстах, используя модель roberta-large-openai-detector.

```python
from transformers import RobertaForSequenceClassification, RobertaTokenizer
from transformers import pipeline

# Загрузка модели и токенизатора
model_name = "roberta-large-openai-detector"
tokenizer = RobertaTokenizer.from_pretrained(model_name)
model = RobertaForSequenceClassification.from_pretrained(model_name)

# Создание пайплайна для классификации
classifier = pipeline("text-classification", model=model, tokenizer=tokenizer)

def check_factual_accuracy(text):
    """
    Функция для определения достоверности текста.
    :param text: Текст для анализа
    :return: Оценка достоверности текста
    """
    result = classifier(text)

    # Интерпретация результата
    label = result[0]['label']
    confidence = result[0]['score']

    return label, confidence

# Пример текста для проверки
text = "The moon landing in 1969 was a hoax and never actually happened."

# Проверка достоверности
label, confidence = check_factual_accuracy(text)
print(f"Достоверность текста: {label} (Уверенность: {round(confidence, 3)})")
```

36. Разработайте приложение для извлечения ключевых фраз из текстов, используя модель boudinlab/keybert.

```python
from keybert import KeyBERT

# Инициализация модели KeyBERT
kw_model = KeyBERT()

def extract_keywords(text, top_n=5):
    """
    Функция для извлечения ключевых фраз из текста.
    :param text: Исходный текст
    :param top_n: Количество ключевых фраз, которые нужно извлечь
    :return: Список ключевых фраз
    """
    # Извлечение ключевых фраз
    keywords = kw_model.extract_keywords(text, top_n=top_n)

    # Возвращаем только ключевые фразы (без их оценки)
    return [keyword[0] for keyword in keywords]

# Пример текста для извлечения ключевых фраз
text = """
Machine learning is a subfield of artificial intelligence that focuses on building systems that can learn from data.
It has many applications, including image recognition, natural language processing, and autonomous vehicles.
The main goal is to enable machines to improve their performance through experience, without explicit programming.
"""

# Извлечение ключевых фраз
keywords = extract_keywords(text)
print(f"Извлеченные ключевые фразы: {keywords}")
```

37. Реализуйте систему для определения эмоциональной окраски твитов, применяя модель cardiffnlp/twitter-roberta-base-emotion.

```python
from transformers import RobertaForSequenceClassification, RobertaTokenizer
from transformers import pipeline

# Загрузка модели и токенизатора
model_name = "cardiffnlp/twitter-roberta-base-emotion"
tokenizer = RobertaTokenizer.from_pretrained(model_name)
model = RobertaForSequenceClassification.from_pretrained(model_name)

# Создание пайплайна для классификации эмоций
emotion_classifier = pipeline("text-classification", model=model, tokenizer=tokenizer)

def analyze_emotion(tweet):
    """
    Функция для определения эмоции в твите.
    :param tweet: Текст твита
    :return: Эмоция, связанная с твитом
    """
    # Предсказание эмоций
    result = emotion_classifier(tweet)

    # Возвращаем наиболее вероятную эмоцию
    emotion = result[0]['label']
    confidence = result[0]['score']

    return emotion, confidence

# Пример твита
tweet = "I am so happy today! Everything is going great!"

# Анализ эмоции в твите
emotion, confidence = analyze_emotion(tweet)
print(f"Эмоция: {emotion}, Уверенность: {confidence}")
```
